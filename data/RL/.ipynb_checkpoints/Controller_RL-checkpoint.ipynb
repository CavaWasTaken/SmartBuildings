```python
import json
import time
import numpy as np
import paho.mqtt.client as mqtt
import stable_baselines3
import os
```

```python
class MyController:
    def __init__(self, TinTrig=26.0, sp_cool_on=22.0, sp_cool_off=28.0,
                 blind_on_status=5, blind_off_status=0):
        """
        TinTrig: soglia di temperatura interna per decidere se "chiudere di più".
        blind_on_status: valore EMS per attivare l'ExteriorBlind (tipicamente 5).
        blind_off_status: valore EMS per disattivare la schermatura (tipicamente 0).
        """
        self.TinTrig = float(TinTrig)
        self.sp_cool_on = float(sp_cool_on)
        self.sp_cool_off = float(sp_cool_off)

        # EMS Control Status values (ShadeStatus)
        self.BLIND_ON = int(blind_on_status)   # proviamo 5 (ExteriorBlind)
        self.BLIND_OFF = int(blind_off_status) # 0 (NoShade)

    def compute_action(self, measurements: dict) -> dict:
        # Letture minime dal payload observation
        Tin_Z1 = float(measurements.get("Tin_Flat2_Zona1", 0.0))
        Tin_Z2 = float(measurements.get("Tin_Flat2_Zona2", 0.0))
        Tin_Z3 = float(measurements.get("Tin_Flat2_Zona3", 0.0))  # non ha shading
        Tin_Z4 = float(measurements.get("Tin_Flat2_Zona4", 0.0))

        Tout = float(measurements.get("Tout", 0.0))
        DNI  = float(measurements.get("DNI", 0.0))

        # (opzionale) energia, solo per log
        Elec = float(measurements.get("Electricity", 0.0))
        DH   = float(measurements.get("DistrictHeating", 0.0))
        DC   = float(measurements.get("DistrictCooling", 0.0))

        # -------------------------
        # Logica "robusta" ON/OFF blind
        # -------------------------
        # Regola base: se c'è sole "significativo" => attiva blind
        # (soglie conservative: puoi ritoccarle dopo)
        blind_base = self.BLIND_ON if DNI >= 150 else self.BLIND_OFF

        # Se zona è calda (oltre TinTrig) tieni blind ON (ma è già ON/OFF)
        # Se vuoi renderla più aggressiva: puoi forzare ON anche con DNI più basso quando Tin > TinTrig
        def blind_cmd(Tin):
            if Tin > self.TinTrig:
                return self.BLIND_ON
            return blind_base

        cmd_Z1 = blind_cmd(Tin_Z1)
        cmd_Z2 = blind_cmd(Tin_Z2)
        cmd_Z4 = blind_cmd(Tin_Z4)

        # debug “raro”
        if not hasattr(self, "_dbg_counter"):
            self._dbg_counter = 0
        self._dbg_counter += 1
        if self._dbg_counter % 60 == 0:
            print(
                f"[CTRL] DNI={DNI:.1f} Tout={Tout:.1f} "
                f"Tin(Z1,Z2,Z3,Z4)=({Tin_Z1:.1f},{Tin_Z2:.1f},{Tin_Z3:.1f},{Tin_Z4:.1f}) "
                f"cmdBLIND(Z1,Z2,Z4)=({cmd_Z1},{cmd_Z2},{cmd_Z4}) "
                f"Elec={Elec:.1f} DH={DH:.1f} DC={DC:.1f}"
            )

        return {
            # FLAT2 - Z1
            "A_Shade_Flat2_Z1_W2": int(cmd_Z1),
            "A_Shade_Flat2_Z1_W8": int(cmd_Z1),
            "A_Shade_Flat2_Z1_W9": int(cmd_Z1),

            # FLAT2 - Z2
            "A_Shade_Flat2_Z2_W2": int(cmd_Z2),
            "A_Shade_Flat2_Z2_W3": int(cmd_Z2),

            # FLAT2 - Z4
            "A_Shade_Flat2_Z4_W2": int(cmd_Z4),

            # STAIRXNOHVAC - Z1 (stesso comando di Z1)
            "A_Shade_Stair_Z1_W2": int(cmd_Z1),
            "A_Shade_Stair_Z1_W3": int(cmd_Z1),
            "A_Shade_Stair_Z1_W4": int(cmd_Z1),
        }
```

```python
# 📦 Buffer per l’ultima osservazione ricevuta
from threading import Lock

obs_buffer = {}
obs_lock = Lock()


# 🧾 MQTT callbacks (solo per il CONTROLLER)
def on_connect_ctrl(client, userdata, flags, rc):
    print(f"✅ Controller — Connected to MQTT broker (rc={rc})")
    client.subscribe("building/observation", qos=1)
    print("✅ Controller — Subscribed to topic: building/observation")


def on_obs_ctrl(client, userdata, msg):
    global obs_buffer
    try:
        payload = msg.payload.decode("utf-8", errors="replace")
        data = json.loads(payload)

        if not isinstance(data, dict):
            print("⚠️ Controller — Unexpected observation payload type:",
                  type(data), data)
            return

        with obs_lock:
            obs_buffer.update(data)

    except Exception as e:
        print(f"⚠️ Controller — Error parsing observation: {e}")


# 🧩 Setup MQTT client
mqtt_broker = os.getenv("MQTT_BROKER_HOST", "mosquitto")
mqtt_port   = int(os.getenv("MQTT_BROKER_PORT", 1883))

client = mqtt.Client()
client.on_connect = on_connect_ctrl
client.on_message = on_obs_ctrl
client.connect(mqtt_broker, mqtt_port)
client.loop_start()


# 🔧 Istanza del controller puro
controller = MyController()
```

```python
# 🔁 Main loop (controller)
try:
    last_log = 0.0

    while True:
        # 1) Snapshot atomico dell'ultima osservazione (e consumo)
        with obs_lock:
            if obs_buffer:
                obs = dict(obs_buffer)   # copia
                obs_buffer.clear()       # consuma
            else:
                obs = None

        # 2) Se non ho osservazioni nuove, aspetto
        if obs is None:
            time.sleep(0.02)
            continue

        # 3) Calcolo azione
        try:
            action_dict = controller.compute_action(obs)
        except Exception as e:
            print(f"[CTRL][ERROR] compute_action failed: {e}")
            time.sleep(0.02)
            continue

        # 4) Pubblico solo se action valida (dict non vuoto)
        if isinstance(action_dict, dict) and action_dict:
            client.publish(
                "building/action",
                json.dumps(action_dict, ensure_ascii=False),
                qos=1
            )

            # log non troppo verboso
            now = time.time()
            if now - last_log > 1.0:
                print(f"[CTRL] Action sent: {action_dict}")
                last_log = now
        else:
            # utile per debug: capisci se compute_action sta tornando vuoto
            now = time.time()
            if now - last_log > 2.0:
                print("[CTRL] No action to send (None/empty).")
                last_log = now

        # 5) piccolo sleep per non saturare CPU
        time.sleep(0.02)

except KeyboardInterrupt:
    print("Stopping controller...")
finally:
    try:
        client.loop_stop()
        client.disconnect()
    except Exception:
        pass
```
