```python

from besos.parameters import Parameter, FieldSelector, FilterSelector, GenericSelector
from besos.parameters import RangeParameter,  CategoryParameter

parameters = []
#------ROOF INSULATION
insulation_r = FieldSelector(class_name='Material', object_name='MW Glass Wool (rolls)_.1445', field_name='Thickness')
roof_range = RangeParameter(min_val = 0.10, max_val=0.50)

insulation_param_roof = Parameter(selector=insulation_r,
                                 value_descriptor=roof_range,
                                 name='Insulation Roof Thickness')
parameters.append(insulation_param_roof)

#-------WALL INSULATION
insulation_w = FieldSelector(class_name='Material', object_name='EPS Expanded Polystyrene (Heavyweight)_.1', field_name='Thickness')
wall_range = RangeParameter(min_val = 0.05, max_val=0.25)

insulation_param_wall = Parameter(selector=insulation_w,
                                 value_descriptor=wall_range,
                                 name='Insulation Wall Thickness')

parameters.append(insulation_param_wall)
#-------WINDOWS

# for elm in building.idfobjects['FENESTRATIONSURFACE:DETAILED']:
#     if elm.Surface_Type == 'Window':
#         wind = FieldSelector(class_name = 'FENESTRATIONSURFACE:DETAILED' , object_name = elm.Name , field_name='Construction_Name')
#         window_categories = CategoryParameter(options=['single_glazing', 'double_glazing' , 'triple_glazing'])
#         window_param =Parameter(selector = wind , value_descriptor= window_categories , name= f'Window_{elm.Name}' )
#         parameters.append(window_param)
#         print(f"  Added parameter for: {elm.Name}")

glazing_cond = FieldSelector(class_name='WindowMaterial:Glazing', object_name='*', field_name='Conductivity')
glazing_cond_range = RangeParameter(min_val = 0.1, max_val=0.8)
glazing_cond_param = Parameter(selector=glazing_cond,
                                 value_descriptors=glazing_cond_range,
                                 name='Glazing - Conductivity')       
# print(glazing_cond_param)

parameters.append(glazing_cond_param)

glazing_visible = FieldSelector(class_name='WindowMaterial:Glazing', object_name='*', field_name='Visible Transmittance at Normal Incidence')
glazing_visible_range = RangeParameter(min_val = 0.01, max_val=0.8)
glazing_visible_param = Parameter(selector=glazing_visible,
                                 value_descriptors=glazing_visible_range,
                                 name='Glazing - Visible')
        
parameters.append(glazing_visible_param)

#the maximum value per both of them in the RangeParameter is 0.8 because you have to check that 
# transmittance + reflectance =< 1 by physical laws, so has some problem with some WINDOWMATERIAL:GLAZING
# that have an high "Back Side Visible Reflectance at Normal Incidence" or "Front Side Visible Reflectance at Normal Incidence"
# the highest in this case was GLASS 38 that has 0.11/0.12 so it's good to have the max_value to 0.8.
# For the project IDF check this.

#-------AIR CHANGES PER HOUR (ACH)
ventAchRange = RangeParameter(min_val=0.0, max_val=5.0)


ventACH_sel = FieldSelector(class_name="ZoneVentilation:DesignFlowRate",
                            object_name="*",
                            field_name="Air Changes per Hour")
ventilationAchParam = Parameter(selector=ventACH_sel,
                                      value_descriptors=ventAchRange,
                                      name="Ventilation ACH")

parameters.append(ventilationAchParam)

#-------SHADING

#range parameters for temperature and radiation shading
TempShadingRange = RangeParameter(min_val=10, max_val=30)
RadShadingRange = RangeParameter(min_val=80, max_val=300)

shade_setpoint_sel = FieldSelector(class_name="WindowShadingControl",
                                   object_name="*",
                                   field_name="Setpoint")
shade_setpoint_param = Parameter(selector=shade_setpoint_sel, 
                                 value_descriptor=TempShadingRange, 
                                 name="Temp Setpoint shading")
parameters.append(shade_setpoint_param)

shade_setpoint2_sel = FieldSelector(class_name="WindowShadingControl",
                                    object_name="*",
                                    field_name="Setpoint_2")
shade_setpoint2_param = Parameter(selector=shade_setpoint2_sel, 
                                  value_descriptor=RadShadingRange, 
                                  name="Rad Setpoint shading")
parameters.append(shade_setpoint2_param)
# print(parameters)
print("\nAll parameters:")
for p in parameters:
    print(f"  - {p.name}")

```

```python
from besos.evaluator import EvaluatorEP
from besos.problem import EPProblem

objectives = ['Electricity:Facility',
              'DistrictCooling:Facility',
              'DistrictHeating:Facility'] # these get made into `MeterReader` or `VariableReader`


problem=EPProblem(parameters, objectives) # problem = parameters + objectives

problem.names()
```

```python
evaluator = EvaluatorEP(problem, 
                        building, 
                        out_dir='outputdir_simo', 
                        err_dir='outputdir_simo',
                        epw=epw_file,
                        progress_bar=True)
```

## Generate the Dataset

1. Generate 10 samples with sampling strategy
2. Setup the parallel processing
3. Simulate the Samples
4. Store and recover the expensive runs

```python
from besos import sampling

num_samples=30
samples = sampling.dist_sampler(sampling.lhs, problem, num_samples=num_samples)
samples
```

Run the samples and calculate the execution time

```python
import time
print("Starting simulations...")
start_time = time.time()
sim_samples = evaluator.df_apply(samples,
                             keep_input=True,  # keep the input data in the results
#                              keep_dirs=True,  
                             processes=4) # flag keep_dirs to True to save all ouput

end_time = time.time()
execution_time = end_time - start_time
```

Calculate the time

```python
def niceformat(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
    return hour, minutes, seconds

hours, mins, secs = niceformat(execution_time)

print(
    "The total running time: {:2.0f} hours {:2.0f} min {:2.0f} seconds".format(
        hours, mins, secs
    )
)
```

# Setup the dataset for the Surrogate Model

```python
y = sim_samples[objectives]
x = sim_samples.drop(columns=objectives)
```

```python
from sklearn import linear_model, pipeline
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, Matern, RationalQuadratic
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
import warnings
'''
window_columns = [col for col in plot_data.columns if col.startswith('Window_')]

glazing_map = {
    'single_glazing': 1,
    'double_glazing': 2,
    'triple_glazing': 3
}

data_output = sim_samples.copy()

for col in window_columns:
    data_output[col] = sim_samples[col].map(glazing_map)
    
y = data_output.drop(features, axis=1)
x = data_output.drop(objectives, axis=1)
'''
train_in, test_in, train_out, test_out = train_test_split(
    x, y, test_size=0.2 , random_state=343420
)

scaler = StandardScaler()
train_in_scale = scaler.fit_transform(X=train_in)
test_in_scale = scaler.fit_transform(X=test_in)

scaler_out = StandardScaler()
train_out_scale = scaler_out.fit_transform(X=train_out)
test_out_scale = scaler_out.fit_transform(X=test_out)
```

Try with more samples 20 and 40

```python
#You can load the data from the files in the sim_samples_folder
# 20 samples generation took: 5 min 18 seconds
# 40 samples generation took: 19 min  7 seconds
```

```python
#GAUSSIAN PROCESS 
hyperparameters = {
    "kernel": [
        None,
        1.0 * RBF(length_scale=1.0, length_scale_bounds=(1e-1, 10.0)),
        1.0 * RationalQuadratic(length_scale=1.0, alpha=0.5),
        # ConstantKernel(0.1, (0.01, 10.0))*(DotProduct(sigma_0=1.0, sigma_0_bounds=(0.1, 10.0))**2),
        1.0 * Matern(length_scale=1.0, length_scale_bounds=(1e-1, 10.0)),
    ]
}

folds = 3

gp = GaussianProcessRegressor(normalize_y=True)

clf = GridSearchCV(gp, hyperparameters, cv=folds)

clf.fit(train_in_scale, train_out_scale)

best_gp = clf.best_estimator_ 

prediction_gp_scaled = best_gp.predict(test_in_scale)
prediction_gp = scaler_out.inverse_transform(prediction_gp_scaled)
print("Predictions:", prediction_gp)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt

print("GP Model MSE:")
mse = mean_squared_error(prediction_gp,test_out.values,multioutput='raw_values',squared=False)/test_out.mean()*100
print(mse)

print("\nGP Model R2 SCORE:")
r2 = r2_score(test_out.values,prediction_gp)
print(r2)
```

```python
#RIDGE REGRESSION

from sklearn.linear_model import RidgeCV


# Opzione 2: RidgeCV per trovare il miglior alpha automaticamente (RACCOMANDATO)
ridge = RidgeCV(alphas=[0.001, 0.01, 0.1, 1.0, 10.0, 100.0], cv=4)
ridge.fit(train_in_scale, train_out_scale)

print(f"Alpha ottimale: {ridge.alpha_}")


# Predizioni su test set (scalate)
prediction_ridge_scaled = ridge.predict(test_in_scale)

# Inverti la scalatura per ottenere i valori reali
prediction_ridge = scaler_out.inverse_transform(prediction_ridge_scaled)

print("\nPredictions:", prediction_ridge)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt

print("RIDGE Model MSE:")
mse = mean_squared_error(prediction_ridge,test_out.values,multioutput='raw_values',squared=False)/test_out.mean()*100
print(mse)

print("\nRIDGE Model R2 SCORE:")
r2 = r2_score(test_out.values,prediction_ridge)
print(r2)
```

## Surrogate Modelling Evaluator object
We can wrap the fitted model in a BESOS `Evaluator`.

```python
from besos.evaluator import EvaluatorEP,EvaluatorGeneric

# Definition of the evaluation function
def evaluation_func(ind, scaler=scaler ):
    """
    Evaluate the input individual using the trained neural network model (nn_model).
    
    Parameters:
    - ind: A single input sample (array-like) to evaluate.
    - scaler: Pre-fitted scaler object for input scaling.
    
    Returns:
    - A list of predicted outputs, scaled back to the original range.
    """
    ind_scaled = scaler.transform([ind])  # Scale the input using the scaler
    prediction = best_gp.predict(ind_scaled)  # Predict using the GP model
    return scaler_out.inverse_transform(prediction)[0].tolist()  # Inverse scale and return as a list

evaluator = EvaluatorGeneric(evaluation_func, problem)

srinputs = sampling.dist_sampler(sampling.lhs, problem, 500)

sroutputs = evaluator.df_apply(srinputs)
srresults = srinputs.join(sroutputs)
srresults.head()
```

# Perform Building Optimization 

Using the best surrogate perfomr an otimization process, selct optimal values and save a new idf with the selected values. Evalute the goodnes of the surrogate simuation with EnergyPlus.

```python
from besos.optimizer import NSGAII

#running NSGA-II optimizator 
results = NSGAII(evaluator, evaluations=5000, population_size=10000)

#plotting results
plt.figure(figsize=(8, 6), dpi=80)
ax = plt.axes(projection='3d')

optres = results.loc[results["pareto-optimal"] == True, :]  # Get only the optimal results

df = pd.DataFrame(optres, columns=features + objectives)

ax.plot3D(results["DistrictCooling:Facility"], 
          results["DistrictHeating:Facility"], 
          results["Electricity:Facility"], 
          "x",
          label="All Results"
         )  # Plot all results in the background as blue crosses

ax.plot3D(optres["DistrictCooling:Facility"], 
          optres["DistrictHeating:Facility"], 
          optres["Electricity:Facility"], 
          "ro",
          label="Pareto-Optimal Points"
         )  # Plot optimal results in red

# Show legend
ax.legend()

ax.set_xlabel("Cooling demand")
ax.set_ylabel("Heating demand")
ax.set_zlabel("Electricity demand")

```

```python

_=pairplot(optres,
               x_vars=samples.columns, 
               y_vars=objectives, kind="scatter")
```

```python
corr = optres.corr()  # Compute the correlation matrix

# Variables to exclude from the correlation heatmap
unuseful = ['violation', 'pareto-optimal']

# Drop unwanted objectives and features from the correlation matrix
corr.drop(objectives + unuseful, axis=1, inplace=True)  # Drop columns
corr.drop(features + unuseful, axis=0, inplace=True)  # Drop rows

# Plotting the refined correlation heatmap
plt.figure(figsize=(13, 10))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Total'] = df['Electricity:Facility'] + df['DistrictHeating:Facility'] + df['DistrictCooling:Facility']
df['Dist'] = df.apply(
    lambda row : 
    np.sqrt(pow(row["DistrictCooling:Facility"],2) + pow(row["DistrictHeating:Facility"],2) + pow(row["Electricity:Facility"],2))
    ,axis=1)

print("Optimal Values:")
df[df.Dist == df.Dist.min()]
```

Identify best building design parameters and save the new idf


```python
optimal_row = df.loc[df['Dist'].idxmin()] 

optimal_params = optimal_row[features].to_dict()

print("Optimal Parameters:", optimal_params)
```
